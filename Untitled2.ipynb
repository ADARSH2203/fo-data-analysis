```python
import polars as pl

df = pl.read_csv(
    "3mfanddo.csv",
    infer_schema_length=0,   # disables aggressive inference
)

print(df.shape)
print(df.dtypes)
```

```python
df = df.with_columns([
    pl.col("STRIKE_PR").cast(pl.Float64),
    pl.col("OPEN").cast(pl.Float64),
    pl.col("HIGH").cast(pl.Float64),
    pl.col("LOW").cast(pl.Float64),
    pl.col("CLOSE").cast(pl.Float64),
    pl.col("SETTLE_PR").cast(pl.Float64),
    pl.col("VAL_INLAKH").cast(pl.Float64),

    #  FIXED columns
    pl.col("CONTRACTS").cast(pl.Float64).cast(pl.Int64),
    pl.col("OPEN_INT").cast(pl.Float64).cast(pl.Int64),
    pl.col("CHG_IN_OI").cast(pl.Float64).cast(pl.Int64),
])
```

```python
df.dtypes
```

```python
df = df.with_columns([
    pl.col("EXPIRY_DT").str.strptime(pl.Date, "%d-%b-%Y", strict=False),
    pl.col("TIMESTAMP").str.strptime(pl.Date, "%d-%b-%Y", strict=False),
])
```

```python
list(zip(df.columns, df.dtypes))
```

```python
df = df.drop_nulls(["OPEN", "CLOSE", "OPEN_INT"])

df = df.unique(subset=[
    "SYMBOL", "EXPIRY_DT", "STRIKE_PR",
    "OPTION_TYP", "TIMESTAMP"
])
```

```python
market_entities = df.select(["SYMBOL", "INSTRUMENT"]).unique()

market_entities = market_entities.with_row_count("entity_id")

market_entities = market_entities.with_columns([
    pl.lit("NSE").alias("exchange"),
    pl.lit(1).alias("exchange_id")
])
```

```python
contract_specs = df.select([
    "SYMBOL", "INSTRUMENT", "EXPIRY_DT",
    "STRIKE_PR", "OPTION_TYP"
]).unique()

contract_specs = contract_specs.join(
    market_entities.select(["entity_id", "SYMBOL", "INSTRUMENT"]),
    on=["SYMBOL", "INSTRUMENT"]
)

contract_specs = contract_specs.with_row_count("contract_id")
```

```python
trading_data = df.join(
    market_entities.select(["entity_id", "SYMBOL", "INSTRUMENT"]),
    on=["SYMBOL", "INSTRUMENT"]
)

trading_data = trading_data.join(
    contract_specs.select([
        "contract_id", "SYMBOL",
        "EXPIRY_DT", "STRIKE_PR", "OPTION_TYP"
    ]),
    on=["SYMBOL", "EXPIRY_DT", "STRIKE_PR", "OPTION_TYP"]
)
```

```python
print(trading_data.shape)

print(
    trading_data.select([
        pl.len(),
        pl.col(["entity_id","contract_id","TIMESTAMP"]).n_unique()
    ])
)
```

```python
trading_data.select([
    pl.len().alias("total_rows"),
    pl.struct(["entity_id", "contract_id", "TIMESTAMP"]).n_unique().alias("unique_rows")
])
```

```python
# Top 10 contracts by Open Interest
trading_data.sort("OPEN_INT", descending=True).select([
    "SYMBOL", "STRIKE_PR", "OPTION_TYP",
    "OPEN_INT", "TIMESTAMP"
]).head(10)
```

```python
# Top 10 contracts by Open Interest
trading_data.sort("CHG_IN_OI", descending=True).head(10)
```

```python
# Price Range Analysis

trading_data = trading_data.with_columns(
    (pl.col("HIGH") - pl.col("LOW")).alias("price_range")
)

trading_data.sort("price_range", descending=True).head(10)
```

```python
# Most Active Symbols (by contracts traded)
trading_data.group_by("SYMBOL").agg(
    pl.sum("CONTRACTS").alias("total_contracts")
).sort("total_contracts", descending=True).head(10)
```

```python
# Expiry-wise activity
trading_data.group_by("EXPIRY_DT").agg([
    pl.sum("CONTRACTS").alias("total_contracts"),
    pl.mean("OPEN_INT").alias("avg_oi")
]).sort("total_contracts", descending=True)
```

```python
# Option Type Analysis (CE vs PE)
trading_data.group_by("OPTION_TYP").agg([
    pl.sum("CONTRACTS"),
    pl.mean("OPEN_INT")
])
```

```python
# Daily trend of OI
trading_data.group_by("TIMESTAMP").agg(
    pl.sum("OPEN_INT").alias("total_oi")
).sort("TIMESTAMP")
```

```python
trading_data.write_csv("final_trading_data.csv")
```

```python
# ================================
# 7. ANALYSIS QUERIES
# ================================

# Q1: Top OI
q1 = trading_data.sort("OPEN_INT", descending=True).head(10)

# Q2: OI Change
q2 = trading_data.sort("CHG_IN_OI", descending=True).head(10)

# Q3: Volatility
trading_data = trading_data.with_columns(
    (pl.col("HIGH") - pl.col("LOW")).alias("price_range")
)
q3 = trading_data.sort("price_range", descending=True).head(10)

# Q4: Most Active Stocks
q4 = trading_data.group_by("SYMBOL").agg(
    pl.sum("CONTRACTS").alias("total_contracts")
).sort("total_contracts", descending=True).head(10)

# Q5: Expiry Analysis
q5 = trading_data.group_by("EXPIRY_DT").agg([
    pl.sum("CONTRACTS").alias("total_contracts"),
    pl.mean("OPEN_INT").alias("avg_oi")
]).sort("total_contracts", descending=True)

# Q6: CE vs PE
q6 = trading_data.group_by("OPTION_TYP").agg([
    pl.sum("CONTRACTS").alias("total_contracts"),
    pl.mean("OPEN_INT").alias("avg_oi")
])

# Q7: Daily Trend
q7 = trading_data.group_by("TIMESTAMP").agg(
    pl.sum("OPEN_INT").alias("total_oi")
).sort("TIMESTAMP")

# ================================
# 8. EXPORT RESULTS
# ================================
q1.write_csv("q1_top_oi.csv")
q2.write_csv("q2_oi_change.csv")
q3.write_csv("q3_volatility.csv")
q4.write_csv("q4_active_stocks.csv")
q5.write_csv("q5_expiry.csv")
q6.write_csv("q6_option_type.csv")
q7.write_csv("q7_daily_trend.csv")

print("All results generated and saved")


```
